### Select columns with methods

```python
movie = pd.read_csv('~/Downloads/movie.csv',
                    index_col='movie_title')
movie.get_dtype_counts()
```

### Use select dtypes method to select only int columns

```python
movie.select_dtypes(include=['int']).head()
```

### Use select dtypes method to select only numeric columns

```python
movie.select_dtypes(include=['number']).head()
```

### Use filter method to select columns that contain Facebook

```python
movie.filter(like='facebook').head()
```

### Search with filter method through regex for all columns that have a digit in their name

```python
movie.filter(regex='\d').head()
```
