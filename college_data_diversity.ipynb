Happens because missing values do not compare equally with one another.

```python
(college_ugds_ == np.nan).sum()
```

Best way to count missing values is isnull() method

```python
college_ugds_.isnull().sum()
```

Get number of non-missing values using count method

```python
college_ugds_.count()
```

You can change axis to 1/columns to count each rows non-missing values

```python
college_ugds_.count(axis='columns').head()
```

You can also sum values in each row. Should add up to 1.

```python
college_ugds_.sum(axis='columns').head()
```

Use median method to get an idea of distribution

```python
college_ugds_.median(axis='index')
```

Determine college diversity

```python
college_ugds_.isnull()\
               .sum(axis=1)\
               .sort_values(ascending=False)\
               .head()
```

Use dropna method to drop all rows that have percentages missing.

```python
college_ugds_ = college_ugds_.dropna(how='all')
college_ugds_.isnull().sum()
```

Use ge method to convert each value to a boolean and see if percentage greater than 15%

```python
college_ugds_.ge(.15)
```

Now use sum to count True values for each college.

```python
diversity_metric = college_ugds_.ge(.15).sum(axis='columns')
diversity_metric.head()
```

Use value_counts to get idea of distribution.

```python
diversity_metric.value_counts()
```

Two schools have more than 15% in five different race categories.

Sort diversity_metric Series to find the two colleges.

```python
diversity_metric.sort_values(ascending=False).head()
```
